# Depth-wise conv test.

```python
import torch
import torch.nn as nn

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# ---------------------------------------------------------
# 1. Standard Convolution (Like ResNet)
# ---------------------------------------------------------
# in_channels=64, out_channels=64, groups=1 (Default)
standard_conv = nn.Conv2d(in_channels=6,
                          out_channels=12, 
                          kernel_size=3, 
                          padding=1, 
                          bias=False)

print("=== STANDARD CONVOLUTION ===")
print(f"Filter Weight Shape: {standard_conv.weight.shape}") 
# Expected Output: [64, 64, 3, 3] -> (Out_Channels, In_Channels, Height, Width)
print(f"Total Parameters:    {count_parameters(standard_conv):,}\n") 
# Expected Output: 36,864


# ---------------------------------------------------------
# 2. Depthwise Convolution (Like NNG)
# ---------------------------------------------------------
# in_channels=64, out_channels=64, groups=64 (The Magic Trigger)
depthwise_conv = nn.Conv2d(in_channels=6, 
                           out_channels=12, 
                           kernel_size=3, 
                           padding=1,
                           groups=6,
                           bias=False)

print("=== DEPTHWISE CONVOLUTION ===")
print(f"Filter Weight Shape: {depthwise_conv.weight.shape}") 
# Expected Output: [64, 1, 3, 3] -> (Out_Channels, In_Channels/Groups, Height, Width)
print(f"Total Parameters:    {count_parameters(depthwise_conv):,}") 
# Expected Output: 576
```

```python
# target output size of 5x7
m = nn.AdaptiveAvgPool2d((5, 7))
input = torch.randn(1, 64, 8, 9)
output = m(input)
print(output.shape)

# target output size of 7x7 (square)
m = nn.AdaptiveAvgPool2d(7)
input = torch.randn(1, 64, 10, 9)
output = m(input)
print(output.shape)

# target output size of 10x7
m = nn.AdaptiveAvgPool2d((None, 7))
input = torch.randn(1, 64, 10, 9)
output = m(input)
print(output.shape)

# target output size of 10x7
m = nn.AdaptiveAvgPool2d(output_size=(1, 1))
input = torch.randn(1, 64, 10, 9)
output = m(input)
print(output.shape)
```
